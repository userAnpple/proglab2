Download csv file for each province (implemente as separate function, which accepts province id as an argument)
source of data https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/vh_adminMean.php?type=Province_Weekly_MeanPlot

```python
import urllib

from datetime import datetime

import pandas as pd
```

```python
def get_csv_province(province_ID, url = "https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?provinceID=1&country=UKR&yearlyTag=Weekly&type=Parea_VHI&TagCropland=crop&year1=1982&year2=2022",
                     file_name = "dataVH"):
    today = datetime.now().strftime("%d_%m_%Y_%H_%M_%S") # get the date and time of downloading

    url_parsed = list(urllib.parse.urlparse(url)) # parse url convert resulting named tuple to list

    query = url_parsed[4].split('&',1) # split query part of parsed url into atomic queries
    query[0] = query[0].split('=',1)[0] + str(province_ID) # modify atomic query provinceID

    url_parsed[4] = '&'.join(query) # reconstruct query part of parsed url

    url = urllib.parse.urlunparse(url_parsed) # reconstruct url

    with urllib.request.urlopen(url) as f:
        data_str = f.read().decode("ASCII")

    data_str = data_str.split("<br>\n<tt><pre>")[1]
    data_str = data_str.replace("</pre></tt>","")
    data_str = data_str.replace(", ",",")
    
    headers = "year,week,"
    for i in range(0, 20):
        headers = headers + "VHI_range_{}_to_{},".format(i * 5, i * 5 + 5)
    headers = headers + "VHI_range_100_to_inf"
        
    data_str = headers + "\n" + data_str
    
    file_name = "{}_{}_{}.csv".format(file_name, province_ID, today) # generate a file name as DataVH_Province_DownloadDate
    
    with open(file_name, "w") as f: # open an existing csv file with corresponding name or create new one with as handles all exceptions
        f.write(data_str) # overwrite or write csv string to aformentioned file
    
    return file_name
```

```python
for i in range(1, 28):
    str_file_name = get_csv_province(i)
    
df = pd.read_csv(str_file_name, index_col = False) # force pandas.read_csv() not to take first column as an index
df.head(5)
```
